```python
import math
import pandas as pd
def task_2_1(P_X_and_Y, V_0):
  P_X = [sum(i for i in P_X_and_Y[j]) for j in range(len(P_X_and_Y[0]))] #Безумовний розподіл P(X) (Сума по рядках)
  print(f'P_X --> {P_X}')
  P_Y = [] #Безумовний розподіл P(Y) (Сума по стовпцях)
  for j in range(len(P_X_and_Y[0])):
    temp = 0
    for i in range(len(P_X_and_Y)):
      temp += P_X_and_Y[i][j]  # Обчислюємо суму по стовпцях
    P_Y.append(round(temp,4))
  print(f'P_Y --> {P_Y}')
  P_Y_after_X = [[0 for _ in range(len(P_X_and_Y[0]))] for _ in range(len(P_X_and_Y))] # Ініціалізація нулевої матриці P(Y|X)
  for i in range(len(P_Y_after_X)):
    for j in range(len(P_Y_after_X[0])):
      P_Y_after_X[i][j] = round(P_X_and_Y[i][j]/P_X[i],4)   #Обчислення кожного елементу матриці
  print('P_Y_after_X = ')
  for i in range(len(P_Y_after_X)):
    print(P_Y_after_X[i])
  H_Y = -sum(i * math.log2(i) for i in P_Y)
  print(f'H_Y --> {H_Y:.4f}')
  H_Y_and_x1 = -sum(i * math.log2(i) for i in P_Y_after_X[0]) #Не заокруглюємо
  print(f'H_Y_and_x1 --> {H_Y_and_x1:.4f}') #Виводимо перші 4 цифри після коми
  I = H_Y - H_Y_and_x1 # Не заокруглюємо
  print(f'I(X,Y) --> {I}')
  C = round(V_0 * (math.log2(len(P_Y)) - H_Y_and_x1),4) #Заокруглюємо
  print(f'C --> {C:.4f}')
  V = round(V_0 * I,4)
  print(f'V --> {V}')

#Запитати, чи заокруглювати, чи ні
P_X_and_Y = [[0.252, 0.175, 0.273],
             [0.078, 0.072, 0.05],
             [0.025, 0.039, 0.036]]
V_0 = 1600

task_2_1(P_X_and_Y,V_0)
```

```python
def task_2_2(p,q,p_b,V_0):
  #Перший спосіб
  C_1 = round(V_0 * ((1-q-p_b)*math.log2(1-q-p_b) + q*math.log2(q) + (1-p_b)*(1 - math.log2(1-p_b))),4) # Формула з презентації
  print(f'C_1 --> {C_1}')
  #Другий спосіб
  P = [[p,p_b,q], # Ініціалізація матриці P
       [q,p_b,p]]
  print('P =')
  for i in P:
    print(i)
  H_Y_after_X = -sum(i * math.log2(i) for i in P[0])
  print(f'H(Y\X) --> {H_Y_after_X}')
  p_x1 = p_x2 = 0.5
  p_y1 = p*p_x1 + q*p_x2
  p_y2 = p_b*p_x1 + p_b*p_x2
  p_y3 = q*p_x1 + p*p_x2
  print(f'p_y1 = {p_y1}',f'p_y2 = {p_y2}',f'p_y3 = {p_y3}')
  H_max_Y = -(p_y1*math.log2(p_y1)+p_y2*math.log2(p_y2)+p_y3*math.log2(p_y3))
  print(f'H_max(Y) --> {H_max_Y}')
  C_2 = round(V_0 * (H_max_Y - H_Y_after_X),4)
  print(f'C_2 --> {C_2}')

task_2_2(0.88,0.06,0.06,2000)
```

```python
def num_method(start,end,step, P_Y_after_X, t = 10**-3): # За умовою завдання з презентації тау = 10^-3 , додав можливість змінювати тау при потребі
  p_x1 = start # Початковий p_x1
  #Ініціалізуємо матриці, що будуть містити результати обрахунків
  C_arr = []
  p_x_arr = []
  p_y_arr = []
  H_Y_arr = []
  H_Y_after_X_arr = []
  while p_x1 < end + step: # Йдемо до end (включно)
    #Формули з презентації
    p_x2 = 1 - p_x1
    p_y1 = p_x1*P_Y_after_X[0][0] + p_x2*P_Y_after_X[1][0]
    p_y2 = p_x1*P_Y_after_X[0][1] + p_x2*P_Y_after_X[1][1]
    print(p_y1,p_y2)
    H_Y_after_x1 = -P_Y_after_X[0][0]*math.log2(P_Y_after_X[0][0]) - P_Y_after_X[0][1]*math.log2(P_Y_after_X[0][1])
    H_Y_after_x2 = -P_Y_after_X[1][0]*math.log2(P_Y_after_X[1][0]) - P_Y_after_X[1][1]*math.log2(P_Y_after_X[1][1])
    H_Y_after_X = p_x1*H_Y_after_x1 + p_x2*H_Y_after_x2
    H_Y = -p_y1*math.log2(p_y1) - p_y2*math.log2(p_y2)
    C = 1/t * (H_Y - H_Y_after_X)
    p_x_arr.append(p_x1)
    p_y_arr.append(p_y1)
    H_Y_arr.append(H_Y)
    H_Y_after_X_arr.append(H_Y_after_X)
    C_arr.append(C)
    p_x1 += step

  df = pd.DataFrame({
      'p(x1)' : p_x_arr,
      'p(y1)' : p_y_arr,
      'H(Y)' : H_Y_arr,
      'H(Y\X)' : H_Y_after_X_arr,
      'C' : C_arr
  })

  print(f'Max C in index # {C_arr.index(max(C_arr))}') # Знаходжу індекс рядка, де C є максимальним. Маючи той рядок, можна легко найти значення для наступно проходу
  return df




P_Y_after_X = [[0.62, 0.38],
               [0.05, 0.95]]

num_method(0.44,0.46,0.001,P_Y_after_X)
```
